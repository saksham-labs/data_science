Exploring seaborn library capabilities

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

%matplotlib inline
```

```python
# df = pd.read_cvs("tips.csv")
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.describe().round()
```

```python
sns.displot(x=tips.total_bill)
plt.show()
```

```python
sns.boxplot(data= tips.total_bill)
```

```python
sns.jointplot(x=tips.total_bill, y=tips.tip, data=tips, kind="scatter")
```

```python
sns.jointplot(x=tips.total_bill, y=tips.tip, data=tips, kind="hex")
```

```python
sns.jointplot(x=tips.total_bill, y=tips.tip, data=tips, kind="reg")
```

```python
#pairplot 
sns.pairplot(tips)
```

```python
#hue is used for categorical variable like male/female 
sns.pairplot(tips,hue="sex")
```

```python
sns.pairplot(tips,hue="sex", palette="binary")
```

```python
#rugplot 
sns.kdeplot(tips.total_bill)
sns.rugplot(x=tips.total_bill)
```

```python
sns.violinplot(x="day", y="total_bill", data=tips )
```

```python
#strip plot 
sns.boxplot(x=tips.day, y=tips.total_bill )
sns.stripplot(x="day", y="total_bill", data=tips, hue="sex" )
```

```python
sns.stripplot(x="day", y="total_bill", data=tips, hue="sex", dodge=True )
```

```python
plt.figure(figsize=(10,5))

sns.violinplot(x="day", y="total_bill", data=tips)
sns.swarmplot(x="day", y="total_bill", data=tips, hue="sex" )
```

**Grids**

```python
iris = sns.load_dataset('iris')
iris.head(3)
```

```python
iris.describe().round()
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
sns.pairplot(iris, hue="species")
```

**Facet grid**

```python
tips=sns.load_dataset('tips')
tips.head(3)
```

```python
# allows us to show histogram divided by categorical varibles --- smoker, lunch time 
g = sns.FacetGrid(tips, col ="time", row='smoker') # creating blank grid 
g.map(plt.hist, "total_bill")
plt.show()
```

```python
#joint grid 
```

```python
g = sns.JointGrid(x="total_bill", y ='tip', data=tips)
g.plot(sns.regplot, sns.histplot)
```
